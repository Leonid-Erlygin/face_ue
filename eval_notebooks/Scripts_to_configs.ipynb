```python
def put_to_quotes(value):
    try:
        value = int(value)
    except Exception as e:
        try:
            value = float(value)
        except Exception as e:
            value = '\"' + value + '\"'
    return value

def arguments_to_yaml(args):
    result = []
    for idx, line in enumerate(args.split("--")):
        print(idx)
#         if "--" not in line:
#             continue
#         line = line.replace("--", "")
        line = line.strip()
        
        if "python" in line:
            continue
        
        
        new_line = None
        if "=" in line:
            arg, value = line.split("=")
            value = put_to_quotes(value)
            new_line = f'{arg}: {value}'
            
        elif "$" in line and "seq" in line:
            line = line.replace("(", "").replace(")", "")
            spl = line.split(" ")
            key = spl[0]
            values = [float(val) for val in spl if type(put_to_quotes(val)) in (int, float)]
            print(f"values={values}")
            new_line = f"{key}: [{values[0]}, {values[2]}, {int((values[2] - values[0]) / values[1])}]"
        
        elif line.count(" ") > 1:
            spl = line.split(" ")
            key, values = spl[0], spl[1:]
            new_line = f'{key}: ['
            
            for val in values:
                new_line = new_line + str(put_to_quotes(val)) + ", "
            new_line = new_line + "]"
        
        elif "device" in line:
            new_line = "device: 0"
        elif "verbose" in line:
            new_line = "verbose: True"
        elif "cached_embeddings" in line:
            new_line = "cached_embeddings: True"
        elif "equal_uncertainty_enroll" in line:
            new_line = "equal_uncertainty_enroll: True"
        
        else:
            raise AssertionError(f"Problem with line : {line}")
            
        result.append(new_line)
        print(f"prev=[{line}]")
        print(f"new=[{new_line}]")
        
    res = "\n".join(result)
    return res

def dump_to_file(string, path):
    with open(path, "w") as f:
        f.write(string)
```
